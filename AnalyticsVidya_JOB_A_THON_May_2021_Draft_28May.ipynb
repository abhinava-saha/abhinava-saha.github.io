# Problem Statement - Credit Card Lead Prediction

Happy Customer Bank is a mid-sized private bank that deals in all kinds of banking products, like Savings accounts, Current accounts, investment products, credit products, among other offerings.



The bank also cross-sells products to its existing customers and to do so they use different kinds of communication like tele-calling, e-mails, recommendations on net banking, mobile banking, etc. 



In this case, the Happy Customer Bank wants to cross sell its credit cards to its existing customers. The bank has identified a set of customers that are eligible for taking these credit cards.



Now, the bank is looking for our help in identifying customers that could show higher intent towards a recommended credit card, given:

- Customer details (gender, age, region etc.)
- Details of his/her relationship with the bank (Channel_Code,Vintage, 'Avg_Asset_Value etc.)

### Check the distribution of data and check for outliers.

```python
# The below function plots a histogram and corresponding boxplot for n variables at a time.
def plot_graph(l, u, df):
    i = 0
    cols = df.select_dtypes(include=[np.number]).columns

    if len(cols[l:u]) == 0:
        return -1
    else:
        j = len(cols[l:u])

    plt.figure(figsize=(5*j, 5))
    f, axs = plt.subplots(2, j, sharex='col',  figsize=(5*j, 5),
                          gridspec_kw={"height_ratios": (0.2, 1), 'wspace': (0.2), 'hspace': (0.00)})
    #(ax_box1,ax_box2,ax_box3), (ax_hist1,ax_hist2,ax_hist3) = axs

    for col in cols[l:u]:
        mean = df[col].mean()
        median = df[col].median()
        mode = df[col].mode()[0]

        if j == 1:
            ax_box = axs[0]
            ax_hist = axs[1]
        else:
            ax_box = axs[0, i]
            ax_hist = axs[1, i]

        sns.boxplot(df[col], ax=ax_box)
        ax_box.axvline(mean, color='r', linestyle='--')
        ax_box.axvline(median, color='g', linestyle='-')
        ax_box.axvline(mode, color='b', linestyle='-')
        ax_box.set_title(col, fontdict={'fontsize': 15, 'color': 'blue'})

        sns.distplot(df[col], ax=ax_hist)
        ax_hist.axvline(mean, color='r', linestyle='--')
        ax_hist.axvline(median, color='g', linestyle='-')
        ax_hist.axvline(mode, color='b', linestyle='-')
        #ax_hist.set_title(col, fontdict={'fontsize': 15, 'color': 'blue'})
        ax_hist.legend({'Mean': mean, 'Median': median, 'Mode': mode})
        i += 1
        ax_box.set(xlabel='')

#    plt.tight_layout()
    plt.show()


# Call the plot_graph function for n variables at a time.
def plot_hist_box(df, n):
    for i in range(0, 100, n):
        x = plot_graph(i, i+n, df)
        if x == -1:
            break
```

### Transformation for Linear Models

## Create a Model Performance Log file to capture the Model performance output.

```python
#Function to create a list of log from the existing model log excel file in working directory.
def create_perflog_list(perf_list, df):
    cols = df.columns
    for row in df.itertuples(index=False): 
        my_dict = {}
        for i, col in enumerate(cols):
            my_dict[col] = row[i]
        perf_list.append(my_dict)

    return perf_list

# Load the Model parameters and performance metrics log file from current working directory, if not present a new one will be created.
import os
from datetime import datetime
wd = os.getcwd()
filename = 'model_perf_log.xlsx'
fpath = wd + '\\' + filename #model log file name (which stores the model parameters and performance metrics)
perf_list = []   #create an empty list.
if os.path.isfile(fpath):
    print("Opening Model Log File at:\n\033[7m{}\033[0m".format(fpath))
    model_perf_log = pd.read_excel(filename)    #read existing log file into dataframe.
    perf_list = create_perflog_list(perf_list, model_perf_log)
else:
    print('The log file \033[7m{}\033[0m does not exist in current working directory. A new log file will be created.'.format(filename))
    print('Current working directory is:\n\033[7m{}\033[0m'.format(wd))
```

```python
def capture_parameters(Model,Predict_On,model_score,AUC,Recall,Precision,F1_Score,best_grid,params,perf_list):
    grid_dict = {}
    grid_dict['Model'] = Model
    grid_dict['Predict_On'] = Predict_On
    grid_dict['Accuracy_Score'] = round(model_score,4)
    grid_dict['Precision'] = round(Precision,4)
    grid_dict['Recall'] = round(Recall,4)
    grid_dict['F1_Score'] = round(F1_Score,4)
    grid_dict['AUC'] = round(AUC,4)
    grid_dict['DateTimeStamp'] = str(datetime.now())
    grid_dict['Best_Parameters'] = str(params)
    grid_dict['Best_Grid'] = str(best_grid)
    
    perf_list.append(grid_dict)
    df = pd.DataFrame(perf_list)   
    return df, perf_list

def generate_model_performance(model_name,thres,predict_type,model,X,y,params,perf_list):
    # Generate the y predicted for Train and Test.
    ypred = model.predict(X)
    if 'SM_Logit' in model_name:
        ypred = sm_predict_y(model, thres, X)
    else:
        ypred = model.predict(X)

    print(f"\033[7m\nAccuracy score of {predict_type} Dataset:\033[0m")  # Accuracy Score
    train_acc = accuracy_score(y, ypred)
    print(f"Accuracy Score of {predict_type} Dataset is: {round(train_acc,4)*100} %")

    print(f"\033[7m\n\nClassification Report of {predict_type} Dataset:\033[0m")  # Classification Reports:
    train_precision, train_recall, train_f1 = classification(y, ypred)
    
    print(f"\033[7m\n\nConfusion Matrix of {predict_type} Dataset:\033[0m")  # Confusion Matrix
    confusion(y, ypred)
    
    print(f"\033[7m\n\nAUC ROC Curve of {predict_type} Dataset:\033[0m") # AUC Score
    train_auc = auc_roc(model_name,model,X,y)
    print(f"\033[7mArea Under the Curve for {predict_type} Dataset is: {round(train_auc,4)*100} %\033[0m\n")
      
    
    model_perf_log, perf_list = capture_parameters(model_name,predict_type,train_acc,train_auc,train_recall,
                                         train_precision,train_f1,model,params,perf_list)
    model_perf_log.to_excel('model_perf_log.xlsx', index=False)
    return model_perf_log, perf_list

def compare_model_performance(model_name,thres,model,X_train,y_train,X_test,y_test,params,perf_list):
    # Generate the y predicted for Train and Test.

    if 'SM_Logit' in model_name:
        ypred_train = sm_predict_y(model, thres, X_train)
        ypred_test = sm_predict_y(model, thres, X_test)
    else:
        ypred_train = model.predict(X_train)
        ypred_test = model.predict(X_test)

    print(f"\033[7m\nAccuracy scores:\033[0m")  # Accuracy Score
    train_acc = accuracy_score(y_train, ypred_train)
    test_acc = accuracy_score(y_test, ypred_test)
    print('Accuracy Score of Training Dataset:',round(train_acc,4)*100,'%')
    print('Accuracy Score of Testing Dataset:',round(test_acc,4)*100,'%')

#    print(f"\033[7m\n\nClassification Reports:\033[0m")  # Classification Reports:
    print(f"\033[7m\n\nClassification Report of Training Dataset:\033[0m")
    train_precision, train_recall, train_f1 = classification(y_train, ypred_train)
    print(f"\033[7m\nClassification Report of Testing Dataset:\033[0m")
    test_precision, test_recall, test_f1 = classification(y_test, ypred_test)
    
#    print(f"\033[7m\nConfusion Matrix Reports:\033[0m")  # Confusion Matrix
    print(f"\033[7m\n\nConfusion Matrix of Training Dataset:\033[0m")
    confusion(y_train, ypred_train)
    print(f"\033[7m\nConfusion Matrix of Testing Dataset:\033[0m")
    confusion(y_test, ypred_test)
    
    print(f"\033[7m\n\nAUC ROC Curve of Training Dataset:\033[0m") # AUC Score
    train_auc = auc_roc(model_name,model,X_train,y_train)
    print('\033[7mArea Under the Curve for Training Dataset',round(train_auc,4)*100,'%\033[0m\n')
    print(f"\033[7m\nAUC ROC Curve of Testing Dataset:\033[0m") # AUC Score
    test_auc = auc_roc(model_name,model,X_test,y_test)
    print('\033[7mArea Under the Curve for Testing Dataset',round(test_auc,4)*100,'%\033[0m\n')
      
   
    model_perf_log, perf_list = capture_parameters(model_name,'Training',train_acc,train_auc,train_recall,
                                         train_precision,train_f1,model,params,perf_list)
    model_perf_log, perf_list = capture_parameters(model_name,'Testing',test_acc,test_auc,test_recall,
                                         test_precision,test_f1,model,params,perf_list)
    model_perf_log.to_excel('model_perf_log.xlsx', index=False)
    return model_perf_log, perf_list
```

```python
def classification(y,ypred):
    print(classification_report(y, ypred, target_names=['Non-Default', 'Default']))
    metrics=classification_report(y, ypred, output_dict=True)
    df=pd.DataFrame(metrics).transpose()
    precision=df.loc["1"][0]
    recall=df.loc["1"][1]
    f1=df.loc["1"][2]
    
    return precision, recall, f1

def confusion(y,ypred):
    cm = confusion_matrix(y,ypred)
    #print(cm)
    classNames = ['Non-Default', 'Default']
    cmlabel = np.array([['TN', 'FP'], ['FN','TP']])
    labels = (np.asarray(["{0} = {1:3d}".format(cmlabel, cmvalue)
                          for cmlabel, cmvalue in zip(cmlabel.flatten(), cm.ravel())])
             ).reshape(2, 2)
    #  'flatten' flattens the array and returns a copy, ravel returns a view of original array.
        #print(labels)
    sns.heatmap(cm, annot=labels, fmt='s', cbar=False,cmap='YlGnBu',
              xticklabels=classNames, yticklabels=classNames)
    plt.xlabel('Predicted Label')
    plt.ylabel('Actual Label')
    plt.title('Confusion Matrix')
    plt.yticks(rotation=0)
    plt.show()
    tn, fp, fn, tp = cm.ravel()
#    print('\nTN={}, FP={}, FN={}, TP={}'.format(tn, fp, fn, tp))
    print('\nActual Positives={}, Predicted Positives={}, Actual Negative={}, Predicted Negative={}'.format(
        (tp+fn),(tp+fp),(tn+fp),(tn+fn)))
    print('\nTotal number of observations = {}'.format(tn+fp+fn+tp))
    
def auc_roc(model_name,model,X,y):
    if 'SM_Logit' in model_name:
        ypred_prob = model.predict(X)    
    else:
        ypred_prob = model.predict_proba(X)[:,1]
        
    auc = roc_auc_score(y,ypred_prob)
    
    fpr, tpr,_=roc_curve(y,ypred_prob)
    plt.plot(fpr, tpr, marker='x', label='')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC')
    plt.show()
    
    return auc

# Function to Predict y for Stats Model Logistic Regression prediction
def sm_predict_y(SM_model, thres, X):
    y_predict_prob = pd.DataFrame(SM_model.predict(X))
    y_class_pred= y_predict_prob.applymap(lambda x: 1 if x>thres else 0)
    return y_class_pred
```
